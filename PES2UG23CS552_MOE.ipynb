```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.5)
```

```python
MODEL_CONFIG = {
    "technical": "You are a Technical Support Expert. Give proper steps so that code is out of bugs.",

    "billing": "You are a Billing Support Expert. Explain charges, refunds, discounts and policies clearly.",

    "general": "You are a General Customer Support Assistant. Be friendly and help customers."
}
```

```python
def route_prompt(user_input):
    # Check for specific keywords first to route to a tool
    if "bitcoin" in user_input.lower():
        return "tool"

    router_prompt = f"""
Classify this query:
[technical, billing, general]

Return ONLY the category name.

Query: {user_input}
"""

    category = llm.invoke(router_prompt).content.strip().lower()

    if category not in MODEL_CONFIG:
        category = "general"

    return category
```

```python
def process_request(user_input):

    category = route_prompt(user_input)

    if category == "tool":
        response = get_bitcoin_price()
    else:
        system_prompt = MODEL_CONFIG[category]

        final_prompt = f"""
{system_prompt}

Customer Query: {user_input}
"""

        response = llm.invoke(final_prompt).content

    return category, response
```

```python
category, answer = process_request(
    "I was charged twice for my subscription"
)

print("Category:", category)
print("Answer:", answer)
```
